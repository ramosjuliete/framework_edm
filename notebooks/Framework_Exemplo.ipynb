# Exemplo de Uso do Framework de Análise SRL

Este notebook demonstra o fluxo **ponta-a-ponta**:
1. Geração do `MinimalDataset`
2. Análise exploratória
3. Clusterização
4. Análise de resultados
5. Execução completa via classe `General`


```python
# Instalação de dependências (se necessário)
!pip install pandas scipy scikit-learn matplotlib hdbscan
```

## 1. Gerar MinimalDataset

```python
from minimaldataset import MinimalDataset

# Exemplo: gerar dataset mínimo a partir de um CSV de eventos brutos
md = MinimalDataset("raw/events.csv", delimiter=";")
md.to_csv("out/minimaldataset.csv")
```

## 2. Análise Exploratória

```python
from exploratoryanalysis import ExploratoryAnalysis

ea = ExploratoryAnalysis(path="out/")
df = ea.loadDataframe("out/minimaldataset.csv", delimiter=";")
ea.statisticalDescription(df)
ea.normalityTeste(df, test="shapiro")
ea.createSpearmanMatrix(df)
```

## 3. Clusterização

```python
from clustering import Clustering

cl = Clustering(path="out/")
df_clusters = cl.kmeans(df, k=3)  # exemplo com KMeans
df_clusters.to_csv("out/df_clusters.csv", index=False)
```

## 4. Análise de Resultados

```python
from resultsanalysis import ResultsAnalysis

ra = ResultsAnalysis(path="out/")
df_clusters = ra.loadFile("out/df_clusters.csv")

ra.statisticalByClusters(df_clusters)
ra.statisticalSignificance(df_clusters)
ra.profileSRLdescription(df_clusters)
ra.profileSRLGraphics(df_clusters)
```

## 5. Execução Completa com `General`

```python
from general import General

g = General(fileDataset="out/minimaldataset.csv", delimiterDataset=";", path="out/")

# Exemplo com Agglomerative
g.execute(algorithm="agglomerative", n_clusters=3, normalityTeste="shapiro")
```
